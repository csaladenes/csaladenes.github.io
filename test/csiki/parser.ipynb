```python
c='Q2'
dx=df.groupby(c).count()[['Response_ID']]
ds=dx*1.0/dx.sum()[0]
ds.columns=['percent']
dx.columns=['count']
dx=dx.join(ds).reset_index().sort_values(by=c,ascending=False)
base=alt.Chart(dx,title=mc[c].split('-')[1].replace(':','').strip())
bar=base.mark_bar(height=20).encode(
    x=alt.X('percent:Q',
            axis=alt.Axis(grid=False,
                title="válaszadók százaléka",
                format=".0%",
                titleAnchor="end",
                titleY=-17,
                titleFontWeight="normal",
                titleFontSize=10),
            #scale=alt.Scale(domain=[0,300])
           ),
    y=alt.Y(c+':N',axis=alt.Axis(title=''),sort=[]),
    color=alt.Color(c+':N',legend=None,scale=alt.Scale(domain=list(dx[c].unique()),
                range=[violet,yellow]))
)
text = bar.mark_text(dx=-5, dy=1, align="right").encode(
    text=alt.Text("percent:N", format=".1%"),
    color=alt.condition(
        "datum['percent'] < 0.08",
        alt.ColorValue(None),
        alt.ColorValue("white"),
    )
    
)
layer = (
    (bar+text)
    .configure_view(stroke=None)
    .configure_title(anchor='start',dx=30)
    .properties(height=alt.Step(25), width=500)
)
layer.save("export/" + c + ".json")
layer.save("export/" + c + ".png")
layer
```
